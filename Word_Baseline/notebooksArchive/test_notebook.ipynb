```python
import numpy as np 
import re
import pandas as pd
import Logger as l

from sklearn.feature_extraction.text import CountVectorizer
```

```python
my_data = pd.read_csv('data\word-freq-top5000.csv', delimiter=',', usecols=[0,1])
my_data
```

```python
D = {str.lower(b):a-1 for a,b in my_data.values}
D
```

```python
def bag_of_words(text):
    # CountVectorizer cf. scikit-learn
    vectorizer = CountVectorizer()

    # Fit transform
    vectorizer.fit_transform([text])

    # Get names
    feature_names = vectorizer.get_feature_names_out()

    # Get the bag of words representation as a dictionary 
    bag_of_word = dict(zip(feature_names, vectorizer.transform([text]).toarray()[0]))

    return bag_of_word

def to_vectorize(text):
    B = bag_of_words(text)
    print(B)
    M = np.zeros(5001)
    for a,b in B.items():
        if a in D:
            M[D[a]] = b
        else:
            # Les mots non présents dans le dictionnaire
            M[5000]+=1
            print(a)
    return M
```
